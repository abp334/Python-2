```python
import socket
host=socket.gethostname()
port=5665
addr=(host,port)
client_socket=socket.socket()
client_socket.connect(addr)
msg=input("Enter request")
while msg!="":
    client_socket.send(msg.encode())
    data=client_socket.recv(1024)
    print("Recieved data :",data.decode())
    msg=input("Enter request")
client_socket.close()
```

```python
import requests
from bs4 import BeautifulSoup
website = "https://subslikescript.com/movie/Titanic-120338"
result = requests.get(website)
soup = BeautifulSoup(result.text,"html.parser")
# print(soup.prettify())
soup.find('h1').text.strip()
script = soup.find('div',class_='full-script').text.strip()
print(script)
with open('titanic.txt','w',encoding='utf-8') as f:
    f.write(script)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
website = "https://www.politifact.com/factchecks/list/"
result = requests.get(website)
soup = BeautifulSoup(result.text,"html.parser")
# print(soup.prettify())
frame = []
links = soup.find_all("li",class_="o-listicle__item")
for i in links:
    statements = i.find("div",class_="m-statement__quote").text.strip()
    link = "https://www.politifact.com"
    link += i.find("div",class_="m-statement__quote").find("a")["href"].strip()
    date = i.find("footer",class_="m-statement__footer").text.split("•")
    date = date[-1].strip()
    source = i.find("a",class_="m-statement__name").text.strip()
    frame.append((statements,link,date,source))
# print(frame)
df = pd.DataFrame(frame,columns=["statements","link","date","source"])
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
path = "Certified used Mercedes-Benz for sale.html"
f = open(path,'r')
response = f.read()
frame = []
soup = BeautifulSoup(response,'html.parser')
l = soup.find_all('div',class_="vehicle-details")
for i in l:
    name = i.find('h2',class_="title").text
    mileage = i.find('div',class_="mileage").text
    dealer = i.find('div',class_="dealer-name").text
    try:
        rating = i.find('span',class_="sds-rating__count").text
    except:
        rating = "NonRated"
    try:
        reviewcount = i.find('span',class_='sds-rating__link sds-button-link').text
    except:
        reviewcount = "None"
    price = i.find('span',class_="primary-price").text
    frame.append((name,mileage,dealer,rating,reviewcount,price))
df = pd.DataFrame(frame,columns=["name","mileage","dealer","rating","reviewcount","price"])
df
```
